```python
import matplotlib.pyplot as plt

floor_l = []
with open('floor_l_file.txt') as file:
    floor_l = file.read().splitlines()
plt.plot(floor_l)
plt.ylabel('floor_l')
plt.show()

floor_r = []
with open('floor_r_file.txt') as file:
    floor_r = file.read().splitlines()
plt.plot(floor_r)
plt.ylabel('floor_r')
plt.show()

prox_l = []
with open('prox_l_file.txt') as file:
    prox_l = file.read().splitlines()
plt.plot(prox_l)
plt.ylabel('prox_l')
plt.show()

prox_r = []
with open('prox_r_file.txt') as file:
    prox_r = file.read().splitlines()
plt.plot(prox_r)
plt.ylabel('prox_r')
plt.show()

plt.scatter(floor_l, floor_r)
plt.show()

plt.scatter(prox_l, prox_r)
plt.show()
```
